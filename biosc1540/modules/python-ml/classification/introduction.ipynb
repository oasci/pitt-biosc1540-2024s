```python
import pandas as pd

df = pd.read_csv("./mushrooms.csv")
print(df[:1])
```

-   **`cap_shape`**: bell = `b`, conical = `c`, convex = `x`, flat = `f`, knobbed = `k`, sunken = `s`
-   **`cap_surface`**: fibrous = `f`, grooves = `g`, scaly = `y`, smooth = `s`
-   **`cap_color`**: brown = `n`, buff = `b`, cinnamon = `c`, gray = `g`, green = `r`, pink = `p`, purple = `u`, red = `e`, white = `w`, yellow = `y`
-   **`bruises`**: bruises = `t`,no = `f`
-   **`odor`**: almond = `a`, anise = `l`, creosote = `c`, fishy = `y`, foul = `f`, musty = `m`, none = `n`, pungent = `p`, spicy = `s`
-   **`gill_attachment`**: attached = `a`, descending = `d`, free = `f`, notched = `n`
-   **`gill_spacing`**: close = `c`, crowded = `w`, distant = `d`

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("white")

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(7, 3), layout="tight")

kwargs = {
    "hue": "class",
    "stat": "probability",
    "multiple": "dodge",
    "shrink": 0.8,
    "palette": {"e": "#277da1", "p": "#f94144"},
}

sns.histplot(data=df, x="cap_shape", ax=axs[0], **kwargs)
sns.histplot(data=df, x="cap_surface", ax=axs[1], **kwargs)

plt.show()
```
