Knowing where money is best spent is crucial for a successful business.
So, if the company sees that spending more on a specific type of ad is linked to higher sales, they can adjust their budget accordingly.
Our idea is to create a model that accurately predicts sales based on the ad budgets.

You may not be able to see any trend just by visually inspecting the graph.
In other words, we can not make any qualitative observations.
We need a quantitative analysis of the data that can give us concrete evidence and confidence in our recommendations.

We can represent this as 

The money spent on ads like TV, social media, and influencer marketing is considered the input, represented as $X_1$, $X_2$, and $X_3$.
The sales, which are influenced by these ad budgets, are the output, $Y$.

If we only had one type advertisement and wanted to compute our return on investment, we could represent this as a linear equation

$$
Y = c + \beta X;
$$

where $\beta$ and $c$ are some unknown numbers.
However, in our case we have

1.  `TV`,
2.  `Billboards`,
3.  `Google_Ads`,
4.  `Social_Media`,
5.  `Influencer_Marketing`,
6.  `Affiliate_Marketing`.

Our expression would then be

$$
Y = c + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3 + \beta_4 X_4 + \beta_5 X_5 + \beta_6 X_6.
$$

If we could figure out what $\beta_1$, $\beta_2$, $\beta_3$, $\beta_4$, $\beta_5$, $\beta_6$ we could know what the business should put more money towards.
Which ever $\beta$ had the largest value would provide the highest rate of return.

However, it gets tiring to write out this whole expression every time, so we use other ways to write it.
Normally when we have a constant like $c$ we just write it as $\beta_0$.
The rest of the terms can be expressed in a $\sum$.

$$
Y = \beta_0 + \sum_{i = 1} \beta_i X_i
$$

Now we can determine these $\beta$ coefficients.
