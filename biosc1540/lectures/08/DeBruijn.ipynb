# De Bruijn



```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def de_bruijn_ize(st, k):
    """Return a list holding, for each k-mer, its left
    k-1-mer and its right k-1-mer in a pair"""
    edges = []
    nodes = set()
    for i in range(len(st) - k + 1):
        edges.append((st[i : i + k - 1], st[i + 1 : i + k]))
        nodes.add(st[i : i + k - 1])
        nodes.add(st[i + 1 : i + k])
    return nodes, edges
```

```python
def visualize_de_bruijn(st, k):
    """Visualize a directed multigraph using networkx and matplotlib"""
    nodes, edges = de_bruijn_ize(st, k)

    G = nx.DiGraph()

    for node in nodes:
        G.add_node(node, label=node)

    for src, dst in edges:
        G.add_edge(src, dst)

    pos = nx.spring_layout(G)
    labels = nx.get_edge_attributes(G, "label")

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_size=700,
        node_color="skyblue",
        font_size=10,
        font_color="black",
        font_weight="bold",
        arrowsize=20,
    )

    plt.show()
```

```python
seq = "ACGCGTCG"
k = 3
nodes, edges = de_bruijn_ize(seq, k)
print(nodes)
print(edges)
visualize_de_bruijn(seq, k)
```

```python
# Generate a longer DNA sequence
seq = "ACGCGTCGAGCTCGATCGTAGCTAGCTCGATCGATC"
k = 3
nodes, edges = de_bruijn_ize(seq, k)
print(seq)
print(nodes)
print(edges)
visualize_de_bruijn(seq, k)
```
